```python
import pandas as pd
import copy
import numpy as np
```

```python
data = pd.read_csv("triplets_all.csv")
data
```

```python
n, N = 878, 385881
```

```python
sz = [0] * n
J = [[0] * n for i in range(n)]
for i in range(N):
    s1, s2 = data.iloc[i, 0], data.iloc[i, 1]
    x1, x2 = map(int, s1.split('_'))
    sz = int(s2)
    if (x1 >= n or x2 >= n):
        break
    J[x1][x2] = sz
    J[x2][x1] = sz
```

```python
def is_clique(graph, subgraph):
    for i in subgraph:
        for j in subgraph:
            if (i != j and graph[i][j] == 0):
                return 0
    return 1

def get_clique(graph):
    del_vertexs = [0] * n
    egraph = copy.deepcopy(graph)
    while (1):
        v = -1
        degs = [0] * n
        for i in range(n):
            cnt = 0
            for j in graph[i]:
                cnt += j > 0
            degs[i] = cnt
        for i in range(n):
            if (del_vertexs[i] == 0 and (v == -1 or degs[v] > degs[i])):
                v = i
        del_vertexs[v] = 1
        for i in range(n):
            graph[i][v] = 0
            graph[v][i] = 0
        clique = []
        
        ver = 0
        sdeg = 1e9
        for i in range(n):
            cnt = 0
            for j in graph[i]:
                cnt += j > 0
            degs[i] = cnt
            if (del_vertexs[i] == 0):
                sdeg = min(sdeg, degs[i])
                ver += 1
        for i in range(n):
            if (del_vertexs[i] == 0):
                clique.append(i)
        if (ver == sdeg + 1 and is_clique(egraph, clique)):
            break
    clique = []
    for i in range(n):
        if (del_vertexs[i] == 0):
            clique.append(i)
    for i in range(n):
        if (del_vertexs[i] == 1):
            ok = 1
            for u in clique:
                if (egraph[u][i] == 0):
                    ok = 0
            if (ok):
                clique.append(i)
    return clique

def greedy():
    res = [[] for i in range(n)]
    color = 0
    while (1):
        graph = [[0] * n for i in range(n)]
        cnt_edges = 0
        for i in range(n):
            for j in range(n):
                if (J[i][j] != 0 and i != j):
                    graph[i][j] = 1 
                    cnt_edges += 1 
        cnt_edges /= 2
        if (cnt_edges == 0):
            return res
        clique = get_clique(graph)
        for i in clique:
            res[i].append(color)
            for j in clique:
                if (i != j):
                    J[i][j] -= 1
        color += 1
        print("C = ", color, "cnt_edges = ", cnt_edges, "clique size = ", len(clique))
    return res
```

```python
greedy()
```
